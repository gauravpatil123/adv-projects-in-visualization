```python
def ghg_plot(dataframe:pd.DataFrame, year_range:list, figsize:tuple=(80, 80)) -> None:
    n_rows = int(len(year_range)/5)
    fig, axes = plt.subplots(n_rows, 5, sharex=False, figsize=figsize)
    fig.suptitle('CO2 Emmisions of top 5 countries by years 1990-2019')

    for year, ax in zip(year_range, axes.flat):
        # col = index % 5
        # row = math.floor(index / 5)
        year = str(year)
        order = dataframe.sort_values(by=[year], ascending=False).iloc[:5]["ISO"]
        sns.barplot(data=dataframe, x="ISO", y=year, order=order, ax=ax, palette="mako")
        ax.set_title(f"Top 5 countries for emmisions in {year}", fontsize=30)
        ax.tick_params(labelrotation=90, labelsize=24)
        ax.set_xlabel("Countries", fontsize=30)
        ax.set_ylabel("CO2 e", fontsize=30)
    plt.subplots_adjust(hspace = 0.3)
```

```python
years = list(range(1990, 2020))
```

```python
ghg_plot(df_ghg_countries, years)
```
